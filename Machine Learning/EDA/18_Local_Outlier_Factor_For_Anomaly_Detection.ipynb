```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles # Fro making Non-Linear Data
```

```python
# Making Dataset
X,Y = make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
plt.scatter(x=X[:,0],y=X[:,1])
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors=6,algorithm='auto',contamination='auto')
labels = lof.fit_predict(X)
```

```python
plt.scatter(x=X[:,0],y=X[:,1],c=labels)
```


##### Hyperparameter tuning for Local Outlier Factor (LOF) involves optimizing its key parameters to improve the performance of outlier detection. Below are the primary parameters to consider and how to tune them effectively.

### Key Parameters to Tune
##### 1.n_neighbors (Number of Neighbors): Determines how many neighbors are used to calculate the local density of a data point.
##### Affects the sensitivity of the algorithm to local density variations.
##### Smaller values detect smaller clusters or anomalies but may result in overfitting.
##### Larger values generalize better but might miss small-scale patterns.

##### 2.contamination (Proportion of Outliers):Expected fraction of outliers in the dataset.
##### Used to set the decision threshold for labeling points as outliers.
##### Proper tuning is essential to avoid misclassifications when the actual number of outliers is unknown.

##### 3.metric (Distance Metric):Determines the distance calculation method for nearest neighbors.
##### Common choices include:
##### Euclidean (default): Straight-line distance.
##### Manhattan: Absolute differences along each dimension.
##### Minkowski: A generalization of Euclidean and Manhattan distances.
##### Custom distance metrics can also be used based on the problem.
