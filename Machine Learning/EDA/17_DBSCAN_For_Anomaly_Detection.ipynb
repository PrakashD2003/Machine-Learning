```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles # Fro making Non-Linear Data
```

```python
# Making Dataset
X,Y = make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
plt.scatter(x=X[:,0],y=X[:,1])
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.1)
dbscan.fit_predict(X)
```

#### How we can identify ouliers in the output of fit_predict of DBSCAN?(All -1 Points are outliers)
##### Cluster Labels in DBSCAN
##### 1.Core Points: Points in dense regions are assigned a cluster label (e.g., 0, 1, 2, ...).
##### 2.Border Points: Points on the edges of dense regions are also assigned cluster labels.
##### 3.Noise Points (Outliers): Points that do not belong to any cluster are labeled as -1 in the output of fit_predict.

#### Tips for Better Outlier Detection
##### 1.Adjust eps: Decrease the eps parameter (neighborhood radius) to make the algorithm more sensitive to noise.
##### 2.Tune min_samples: Increase min_samples to require more neighbors for a point to be considered a core point.
##### 3.Cluster Validity Measures: Use metrics like the Silhouette Score to validate the clustering and ensure that noise points are well-separated.

```python
plt.scatter(x=X[:,0],y=X[:,1],c=dbscan.labels_)
```
