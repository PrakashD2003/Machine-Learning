```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons ## For creating Non-linear Dataset
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
## Difference between DBSCAN and K-Means is that DBSCAN acan also be used fro Non-Linear data
# Creating Non-Linear Dataset
X,Y = make_moons(n_samples=250,noise=0.05)
```

```python
plt.scatter(x=X[:,0],y=X[:,1])
```

```python
## Standardization
X_train,X_test,Y_train,Y_test = train_test_split(X,X,test_size=0.25,random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
## DBSCAN
from sklearn.cluster import DBSCAN

dbscan_model =DBSCAN(eps=0.5)
dbscan_model.fit(X_train_scaled)
```

```python
dbscan_model.labels_
```

```python
plt.scatter(X_train_scaled[:,0],X_train_scaled[:,1],c=dbscan_model.labels_)
```
